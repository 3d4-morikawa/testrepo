**Read the CSV File into a Pandas DataFrame**

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
# SQLiteデータベースに接続
conn = sqlite3.connect('survey-data.sqlite')  # 例: 'survey_data.db'

# SQLクエリで CompTotal を取得
query = "SELECT CompTotal FROM main"
df = pd.read_sql_query(query, conn)

# 欠損値を除外し、異常値（上位5%）を除外して視認性を向上
df_clean = df['CompTotal'].dropna()
df_filtered = df_clean[df_clean < df_clean.quantile(0.95)]

# ヒストグラムを描画
plt.figure(figsize=(10, 6))
plt.hist(df_filtered, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Compensation (CompTotal)')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
# 年齢カテゴリを取得
query = "SELECT Age FROM main"
df = pd.read_sql_query(query, conn)

# 欠損除外 & 集計
age_counts = df['Age'].dropna().value_counts().sort_index()

# 横棒グラフで表示
plt.figure(figsize=(10, 6))
plt.barh(age_counts.index, age_counts.values, color='lightgreen')
plt.title('Respondent Count by Age Group')
plt.xlabel('Number of Respondents')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# 年齢カテゴリを中央値に変換するマップ
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# SQLite に接続
conn = sqlite3.connect('survey-data.sqlite')

# Age と WorkExp を取得
query = "SELECT Age, WorkExp FROM main"
df = pd.read_sql_query(query, conn)

# Age を数値に変換（カテゴリ → 中央年齢）
df['Age_numeric'] = df['Age'].map(age_map)

# WorkExp を数値に変換（必要なら）
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# 欠損を除外
df_clean = df[['Age_numeric', 'WorkExp']].dropna()

# 散布図を描画
plt.figure(figsize=(8, 6))
plt.scatter(df_clean['Age_numeric'], df_clean['WorkExp'], alpha=0.6, color='steelblue', edgecolors='k')
plt.title('Scatter Plot of Age vs Work Experience')
plt.xlabel('Age (converted from age group)')
plt.ylabel('Work Experience (Years)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
# 必要な列を取得
query = "SELECT TimeSearching, Frustration, Age FROM main"
df = pd.read_sql_query(query, conn)

# カテゴリ形式の Age → 数値に変換
df['AgeNumeric'] = df['Age'].map(age_map)

# 数値に変換（文字列などが入っていた場合に備える）
df['TimeSearching'] = pd.to_numeric(df['TimeSearching'], errors='coerce')
df['Frustration'] = pd.to_numeric(df['Frustration'], errors='coerce')

# 欠損を除去
df_clean = df[['TimeSearching', 'Frustration', 'AgeNumeric']].dropna()

# NumPy配列に変換（安全のため）
x = df_clean['TimeSearching'].values
y = df_clean['Frustration'].values
sizes = df_clean['AgeNumeric'].values * 3  # バブルサイズスケーリング

# バブルプロット描画
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, c='skyblue', edgecolors='gray', linewidth=0.5)
plt.title('Bubble Plot: TimeSearching vs Frustration (Bubble Size = Age)')
plt.xlabel('Time Spent Searching for Information')
plt.ylabel('Frustration Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
# データベースの希望列を取得
query = "SELECT DatabaseWantToWorkWith FROM main"
df = pd.read_sql_query(query, conn)

# 欠損を除外し、カンマで分割 → フラットなリストに展開
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(',')

# 空白を除去しながら1つのリストに統合
all_dbs = [db.strip() for sublist in db_series for db in sublist]

# 頻度をカウントして上位5件を取得
top5 = Counter(all_dbs).most_common(5)

# ラベルと値に分割
labels = [item[0] for item in top5]
sizes = [item[1] for item in top5]

# 円グラフを描画
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Work With Next Year')
plt.tight_layout()
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
# 必要な列を取得
query = "SELECT Age, TimeSearching, TimeAnswering FROM main"
df = pd.read_sql_query(query, conn)

# 年齢カテゴリを中央値に変換（あとでラベル用にも使用）
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# 年齢カテゴリを数値に変換
df['Age_numeric'] = df['Age'].map(age_map)

# TimeSearching & TimeAnswering を数値化
df['TimeSearching'] = pd.to_numeric(df['TimeSearching'], errors='coerce')
df['TimeAnswering'] = pd.to_numeric(df['TimeAnswering'], errors='coerce')

# 年齢が 30〜35歳 に相当するカテゴリ：29.5〜39.5
df_filtered = df[(df['Age_numeric'] >= 30) & (df['Age_numeric'] <= 35)][['TimeSearching', 'TimeAnswering']].dropna()

# 中央値を算出
median_searching = df_filtered['TimeSearching'].median()
median_answering = df_filtered['TimeAnswering'].median()

# 積み上げ棒グラフを描画
plt.figure(figsize=(6, 6))
plt.bar('Age 30–35', median_searching, label='Time Searching')
plt.bar('Age 30–35', median_answering, bottom=median_searching, label='Time Answering')

plt.title('Median TimeSearching and TimeAnswering (Age 30–35)')
plt.ylabel('Median Time (minutes)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
# Age と CompTotal を数値に変換
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

# 45〜60歳にフィルタリング & 欠損除外
df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)][['Age', 'CompTotal']].dropna()

# 年齢ごとの CompTotal の中央値を計算
median_by_age = df_filtered.groupby('Age')['CompTotal'].median()

# プロット
plt.figure(figsize=(10, 6))
plt.plot(median_by_age.index, median_by_age.values, marker='o')
plt.title('Median Total Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Total Compensation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
# 欠損を除外し、MainBranchの値をカウント
branch_counts = df['MainBranch'].dropna().value_counts()

# 横棒グラフを描画
plt.figure(figsize=(10, 6))
plt.barh(branch_counts.index, branch_counts.values, color='skyblue')
plt.title('Distribution of Respondents by MainBranch')
plt.xlabel('Number of Respondents')
plt.ylabel('MainBranch')
plt.tight_layout()
plt.show()
```

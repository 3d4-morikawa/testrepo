##### Step 1: Install and Import Required Libraries

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Figure size
plt.figure(figsize=(12, 6))

# Histogram with KDE (Kernel Density Estimate)
sns.histplot(df['ConvertedCompYearly'], bins=50, kde=True)

# Titles and labels
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('Yearly Compensation')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
# フルタイム雇用者のデータをフィルタリング
full_time_df = df[df['Employment'] == "Employed, full-time"]

# 中央値を計算
median_comp = full_time_df['ConvertedCompYearly'].median()

# 結果を表示
print(f"Median Yearly Compensation for Full-Time Employees: {median_comp}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 上位の国（データ数が多い国）を抽出
top_countries = df['Country'].value_counts().head(10).index
df_filtered = df[df['Country'].isin(top_countries)]

# ボックスプロットを作成
plt.figure(figsize=(14, 7))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_filtered)

# 軸のラベルとタイトル
plt.xticks(rotation=45)  # 国名を見やすく回転
plt.yscale('log')  # 対数スケールで表示（極端な外れ値の影響を抑える）
plt.title('Distribution of Yearly Compensation Across Countries')
plt.xlabel('Country')
plt.ylabel('Yearly Compensation (log scale)')

# グラフを表示
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
# IQR（四分位範囲）を計算
Q1 = df['ConvertedCompYearly'].quantile(0.25)  # 第1四分位数
Q3 = df['ConvertedCompYearly'].quantile(0.75)  # 第3四分位数
IQR = Q3 - Q1  # 四分位範囲

# 外れ値の定義（1.5×IQRの範囲外のデータを除外）
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 外れ値を除外した新しいDataFrameを作成
df_refined = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# 結果を表示（新しいデータの行数を確認）
print(f"Original dataset size: {len(df)}")
print(f"Refined dataset size: {len(df_refined)}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
# 必要な列を選択（外れ値除去後のデータを使用する場合は df_refined を適用）
correlation_df = df_refined[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']]

# 相関行列を計算
correlation_matrix = correlation_df.corr()

# ヒートマップをプロット
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# タイトルを設定
plt.title('Correlation Heatmap: Compensation, Work Experience, and Job Satisfaction')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
# 図のサイズ設定
plt.figure(figsize=(14, 6))

# 1つ目の散布図: ConvertedCompYearly vs WorkExp
plt.subplot(1, 2, 1)  # 1行2列の1つ目
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=df_refined, alpha=0.5)
plt.title('Scatter Plot: Work Experience vs Yearly Compensation')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Yearly Compensation')

# 2つ目の散布図: ConvertedCompYearly vs JobSatPoints_1
plt.subplot(1, 2, 2)  # 1行2列の2つ目
sns.scatterplot(x='JobSatPoints_1', y='ConvertedCompYearly', data=df_refined, alpha=0.5)
plt.title('Scatter Plot: Job Satisfaction vs Yearly Compensation')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Yearly Compensation')

# グラフを表示
plt.tight_layout()  # レイアウト調整
plt.show()
```

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# 欠損を除去して分割・展開
df_clean = df.dropna(subset=['DatabaseWantToWorkWith'])
df_clean['Database'] = df_clean['DatabaseWantToWorkWith'].str.split(',')
df_clean = df_clean.explode('Database')
df_clean['Database'] = df_clean['Database'].str.strip()

# データベースごとの出現数をカウント
db_counts = df_clean['Database'].value_counts()

# 上位5件＋その他をまとめる
top5 = db_counts.head(5)
others = db_counts[5:].sum()
top5['Other'] = others

# 円グラフの描画
plt.figure(figsize=(8, 8))
plt.pie(
    top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 Most-Desired Databases (DatabaseWantToWorkWith)')
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# 欠損除去とカンマ区切り展開
df_clean = df.dropna(subset=['DevType'])
df_clean['DevRole'] = df_clean['DevType'].str.split(',')
df_clean = df_clean.explode('DevRole')
df_clean['DevRole'] = df_clean['DevRole'].str.strip()

# 出現頻度をカウント
role_counts = df_clean['DevRole'].value_counts()

# 上位5件＋その他をまとめる
top5_roles = role_counts.head(5)
others = role_counts[5:].sum()
top5_roles['Other'] = others

# 円グラフ描画
plt.figure(figsize=(8, 8))
plt.pie(
    top5_roles.values,
    labels=top5_roles.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Developer Roles (DevType)')
plt.tight_layout()
plt.show()
```
